```python
#Our review rating colomn has some missing values so the solution if to find the median rating by category and use that to pill in our missing values
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Changing the column naming format
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a column age_group
labels = ['Young Adult', 'Adult', 'Middle_aged', 'Senior']
df['age_group']= pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']]
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
# create a column purshase_frequency_days
freguency_mapping = {
    'Every 3 Months' :90,   
    'Annually' : 365,         
    'Quarterly' :90,        
    'Monthly' :30,          
    'Bi-Weekly' :14,        
    'Fortnightly' :14,      
    'Weekly' :7,         
}

df['frequency_purchases_ days'] = df['frequency_of_purchases'].map(freguency_mapping)
```
